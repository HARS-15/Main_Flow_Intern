# Data Cleaning and Preprocessing

### Importing Libraries and Loading data

```python
import pandas as pd
import numpy as np
data=pd.read_csv("data/01.Data Cleaning and Preprocessing.csv")
```

### Data View

```python
data.head()

k=data.shape
print("No of features:",k[1])
print("No of objects:",k[0])
```

### Managing Duplicates

```python
data.drop_duplicates(inplace=True)
k=data.shape
print("No of features:",k[1])
print("No of objects:",k[0])
```

### Statistical descripiction

```python
data.describe()
```

### Null values

```python
data_null=data.isnull().sum()
print(data_null)
```

#### Modifing Null values 

```python
data_modified1=data.fillna(0)
data_null=data_modified1.isnull().sum()
print(data_null)
```

#### Forwarding Filling

```python
data_modified=data.fillna(method="pad")
data_null=data_modified.isnull().sum()
print(data_null)
```

#### Backward Filling

```python
data_modified=data.fillna(method="bfill")
data_null=data_modified.isnull().sum()
print(data_null)
```

#### Dropping null values

```python
data_cleaned=data.dropna()
k=data_cleaned.shape
print("No of features:",k[1])
print("No of objects:",k[0])
```

### Data Features

```python
data.columns

#droping a particular feature
data.drop(["BlowFlow"],inplace=True,axis=1)
print(data.columns)
```

### Convertion of datatype

```python
data_modified1 = data_modified1.apply(pd.to_numeric, errors='coerce')
```

### Inter Quartile Range

```python
Q1 = data_modified1.quantile(0.25)
Q3 = data_modified1.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
filtered_data = data_modified1[~((data_modified1 < lower_bound) | (data_modified1 > upper_bound)).any(axis=1)]
print(filtered_data.shape)
```
